
Renaming Columns


```python
data.rename(columns = {"Entity": "Country", "Maize Production (tonnes)": "Maize (tonnes)", 
                        "Rice  Production ( tonnes)": "Rice (tonnes)", "'Yams  Production (tonnes)": "Yams (tonnes)",
                        "Wheat Production (tonnes)": "Wheat (tonnes)", "Tomatoes Production (tonnes)": "Tomatoes (tonnes)",
                        "Tea  Production ( tonnes )" : "Tea (tonnes)", "Sweet potatoes  Production (tonnes)": "Sweet potatoes (tonnes)", 
                        "Sunflower seed  Production (tonnes)": "Sunflower_seed (tonnes)", "Sugar cane Production (tonnes)": "Sugar_cane (tonnes)", 
                        "Soybeans  Production (tonnes)": "Soybeans (tonnes)", "Rye  Production (tonnes)": "Rye (tonnes)", 
                        "Potatoes  Production (tonnes)": "Potatoes (tonnes)", "Oranges  Production (tonnes)": "Oranges (tonnes)",
                        "Peas, dry Production ( tonnes)": "Peas (tonnes)",  "Palm oil  Production (tonnes)": "Palm_oil (tonnes)",
                        "Grapes  Production (tonnes)": "Grapes (tonnes)", "Coffee, green Production ( tonnes)": "Coffee (tonnes)", 
                        "Cocoa beans Production (tonnes)": "Cocoa_beans (tonnes)", "Meat, chicken  Production (tonnes)": "Meat (tonnes)",
                        "Bananas  Production ( tonnes)" : "Bananas (tonnes)", "Avocados Production (tonnes)" : "Avocados (tonnes)",
                        "Apples Production (tonnes)" : "Apples (tonnes)"}, inplace = True)
```

```python
# Have to use this approach to rename the column again

data.columns = [col.replace('( tonnes )', '(tons)') for col in data.columns]
data.columns = [col.replace(' Production ( tonnes)', '') for col in data.columns]
data.columns = [col.replace('Tea  Production (tons)', 'Tea (tons)') for col in data.columns]
data.columns = [col.replace('Meat, chicken  (tons)', 'Meat (tons)') for col in data.columns]
data.columns = [col.replace('Bananas (tons)(tons', 'Bananas (tons)') for col in data.columns]
data.columns = [col.replace('Coffee, green', 'Coffee(tons)') for col in data.columns]
data.columns = [col.replace('Peas, dry', 'Peas (tons)') for col in data.columns]
data.columns = [col.replace('Sunflower seed  (tons)', 'Sunflower seed (tons)') for col in data.columns]
data.columns = [col.replace('Soybeans  (tons)', 'Soybeans (tons)') for col in data.columns]
data.columns = [col.replace('Rye  (tons)', 'Rye (tons)') for col in data.columns]
data.columns = [col.replace('Potatoes  (tons)', 'Potatoes (tons)') for col in data.columns]
data.columns = [col.replace('Oranges  (tons)', 'Oranges (tons)') for col in data.columns]
data.columns = [col.replace('Palm oil  (tons)', 'Palm oil (tons)') for col in data.columns]
data.columns = [col.replace('Grapes  (tons)', 'Grapes (tons)') for col in data.columns]
data.columns = [col.replace('Coffee(tons)', 'Coffee (tons)') for col in data.columns]
data.columns = [col.replace('Rice (tons)(tons)', 'Rice (tons)') for col in data.columns]
data.columns = [col.replace('Yams  (tons)', 'Yams (tons)') for col in data.columns]
data.columns = [col.replace('Sweet potatoes  (tons)', 'Sweet potatoes (tons)') for col in data.columns]


```
